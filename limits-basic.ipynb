```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
def fn(my_x):
    my_y = 1/my_x
    return my_y
```

```python
print (fn(-1))
print (fn(-0.1))
print (fn(-0.01))
```

```python
y = fn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axvline(x=0, color='purple', linestyle='--')
plt.axhline(y=0, color='purple', linestyle='--')
_ = ax.plot(x,y)
```

# Schaum's Outline: Calculus

This following are from the book of Schaum's outline: Calculus

```python
# 17 pg. 63
# (a)

def fn(x_a):
    y_a = (7*x_a**9 - 4*x_a**5 + 2*x_a - 13)/(-3*x_a**9 + x_a**8 - 5*x_a**2 + 2*x_a)
    return y_a
```

```python
y = fn(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axvline(x=0, color='red', linestyle='--')
plt.axhline(y=-(7/3), color='red', linestyle='--')
_ = ax.plot(x,y)
```
